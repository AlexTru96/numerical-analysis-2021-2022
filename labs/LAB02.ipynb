```python
def interpolate(q,x,function):
    """Use the q points to construct interpolation of function on the points x.
        Return the polynomial interpolation function evaluated at x
    """
    n=len(q)
    V = array([q**j for j in range(n)]).T
    y = array([x**j for j in range(n)]).T

    r= function(q)
    p= linalg.solve(V,r) # Vinverse.dot(function(q))

    return y.dot(p)
```

```python

interpolate(q,x,runge)
```

```python

plot(x,runge(x))
plot(x,interpolate(q,x,runge))
plot(q,runge(q),'ro')
```
